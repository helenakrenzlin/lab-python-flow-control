```python
#1 
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory ={}
print (products)
```

```python
for item in products:
    value = int(input(f"Enter the quantity for {item}: "))
    inventory[item] = value
print (inventory)
```

```python
customer_orders = set()
```

```python
for item in range (3):
    product_ordered = input(f"Enter the name of the product {item +1} ")
    customer_orders.add(product_ordered)
print (customer_orders)
```

```python
#2
continue_ordering = "yes"
while continue_ordering == "yes" :
    products_ordered = input(f"Enter the the name of the product:")
    customer_orders.add(products_ordered)
    continue_ordering =input(f"Do you want to continuing ordering (yes/no)?")
    
print (f"Final customer orders:{customer_orders}")
```

```python
#3
for ordered_product in customer_orders: 
    if ordered_product in inventory: 
        inventory[ordered_product]-= 1
print (f"Updated Inventory: {inventory}")
```
